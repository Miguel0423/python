```python
# Operacione Matematicas
#4to nivel de importancia
# suma ordinaria
print(3 + 4)
numero1,numero2 = 12,25
print(numero1 + numero2)

#Resta Ordinaria
print(3 - 4)

#Declaracion de Variable
v1 = 3
v2 = 12
resultado = v1- v2
print(resultado)
```

```python
#3er Nivel de importancia
# Multiplicacion Ordinaria
M1, M2 = 2,3.14
print(M1 * M2) # 6.28

# DIvision Ordinaria
M1,M2,M3,M4,M5,M6 = 2,5,6,5,4,5
print(M1 / M2) # 0.4
print(M3 / M4) # 0
print(M5 / M6) # 1.

# Division de Piso
print(M1 // M2) # Lo redondea mass cercano asia el cero
print(M3 // M4)
print(M5 // M6)
```

```python
# Potencia
print(2 ** 3) # 8
P1,P2 = 7,2
print(P1 ** P2) # 49
```
